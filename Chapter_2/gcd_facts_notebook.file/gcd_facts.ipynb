### ðŸ”­ An example

> **Theorem** If $\gcd(a,b)=1$, then $\gcd(a+b,a-b)=1$ or $2$.

Before starting on the proof, let's check what we can find.

```python
coprime_pairs = [(a,b) for a in range(1,5) for b in range(a,6) if gcd(a,b)==1]

print("   a    b    gcd(a+b,a-b)")
print("---------------------------")
for a,b in coprime_pairs:
    print(f"{a:>4} {b:>4} {gcd(a+b,a-b):^13}")            
```

And now that we have some experimental data, we attempt a proof.

**Proof** Let $a,b$ be integers with $\gcd(a,b)=1$. Let $d=\gcd(a+b,a-b)$ and take $x,y\in\mathbb{Z}$ such that $dx=a+b$ and $dy=a-b$.  Combining these two equations we find
$$d(x+y)=2a\quad\text{and}\quad d(x-y)=2b.$$
We conclude that $d$ divides both $2a$ and $2b$ so it divides $\gcd(2a,2b)=2\gcd(a,b)=2$.  And thus the only possible values for $d$ are $1$ and $2$.

<div style="text-align: right">ðŸ”²</div>

The computation also provided something else interesting: 2 only happens then $a$ and $b$ are both odd. So let's write this as a formal result that we can prove.

> **Corollary** If $\gcd(a,b)=1$, then $\gcd(a+b,a-b)=2$ if and only if $a$ and $b$ are both odd.

**Proof** 
($\Rightarrow$) Assume $\gcd(a+b,a-b)=2$, then $a+b$ and $a-b$ are both even. This only happens if $a$ and $b$ have the same parity (they are both even or both odd).  Note that $a$ and $b$ cannot both be even, as that would imply $2\mid\gcd(a,b)=1$.  Thus $a$ and $b$ must be odd.

($\Leftarrow$) If $a$ and $b$ are both odd, then $a+b$ and $a-b$ are even, so $2\mid \gcd(a+b,a-b)$.  From the theorem, we know that $\gcd(a+b,a-b)=1$ or $2$. We conclude that $\gcd(a+b,a-b)=2$.
